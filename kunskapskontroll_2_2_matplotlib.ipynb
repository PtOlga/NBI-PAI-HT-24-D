### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure. The whole figure. The Figure keeps track of all the child Axes, 
# a group of 'special' Artists (titles, figure legends, colorbars, etc.), and even nested subfigures.

# An Axes is an Artist attached to a Figure that contains a region for plotting data, and usually includes two or three Axis objects 
# that provide ticks and tick labels to provide scales for the data in the Axes.

# Axis set the scale and limits and generate ticks (the marks on the Axis) and ticklabels (strings labeling the ticks).     

# Everything visible on the Figure is an Artist 
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 10)
y = x**2

# Object-Oriented 
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, y, label="y = x^2", color="red")
ax.set_title("Object-oriented approach")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend()

# Pyplot 
plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, y, label="y = x^2", color="blue")
plt.title("Pyplot approach")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot
axs[0].scatter(random_data_x, random_data_y, alpha=0.5, color='blue', label='Random data')
axs[0].plot(x, y, color='red', label='y = x^2')
axs[0].set_title("Scatter Plot")
axs[0].set_xlabel("Random X")
axs[0].set_ylabel("Random Y")
axs[0].legend()

# Bar plot
axs[1].bar(names, values, color='orange')
axs[1].set_title("Bar Plot")
axs[1].set_xlabel("Fruit")
axs[1].set_ylabel("Quantity")

# Adjust layout and show plot
plt.tight_layout()
```
